```python
# Função para baixar e armazenar dados
def processar_dia(data):
    # Formatando a data para a URL
    earth_date = data.strftime('%Y-%m-%d')
    url = f"https://api.nasa.gov/mars-photos/api/v1/rovers/Perseverance/photos?earth_date={earth_date}&api_key={api_key}"

    # Obtendo os dados da API
    dados = obter_dados_api(url)

    if dados:
        # Criar o diretório para o dia
        day_directory = os.path.join(main_directory, earth_date)
        os.makedirs(day_directory, exist_ok=True)

        # Inicializando uma lista para armazenar os dados do rover
        photo_data = []
        for photo in dados['photos'][:10]:  # Limitar a 10 fotos
            photo_data.append({
                'photo_id': photo['id'],
                'sol': photo['sol'],
                'camera_id': photo['camera']['id'],
                'camera_name': photo['camera']['name'],
                'camera_full_name': photo['camera']['full_name'],
                'img_src': photo['img_src'],
                'earth_date': photo['earth_date'],
                'rover_id': photo['rover']['id'],
                'rover_name': photo['rover']['name'],
                'rover_landing_date': photo['rover']['landing_date'],
                'rover_launch_date': photo['rover']['launch_date'],
                'rover_status': photo['rover']['status'],
            })

        # Criando o DataFrame
        df = pd.DataFrame(photo_data)

        # Salvar o DataFrame em formato CSV
        df.to_csv(os.path.join(day_directory, 'nasa_photos.csv'), index=False)

        # Baixar cada uma das fotos e salvar no diretório
        for photo in photo_data:
            img_url = photo['img_src']
            img_name = os.path.join(day_directory, f"{photo['photo_id']}.jpg")
            response = requests.get(img_url)
            if response.status_code == 200:
                with open(img_name, 'wb') as img_file:
                    img_file.write(response.content)
                    print(f"imagem a imagem: {img_name}")

# Loop para processar todos os dias do mês de abril
ano = 2025
for dia in range(1, 30):
    data_atual = datetime(ano, 4, dia)
    processar_dia(data_atual)
```
